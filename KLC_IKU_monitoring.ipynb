```
import requests
from bs4 import BeautifulSoup
import re
import datetime

tgl_hari_ini = datetime.date.today().strftime("%d-%m-%Y")

def cek_klc(url):
    s = requests.Session()
    r = s.get(url)
    if r.status_code == 200:
        bsoup = BeautifulSoup(r.text, 'html.parser')
        rating = re.search(r"Rating: <strong.*?>(.+?)</strong>", r.text).group()
        rating = re.search("\d+\.\d+", rating).group()
        votes = re.search(r"From \d+ votes", r.text).group()
        votes = re.search(r"\d+", votes).group()
        views = re.search(r"Views : \d+", r.text).group()
        views = re.search(r"\d+", views).group()
        results = [url, rating, votes, views]
        return results
   
if __name__ == "__main__":
    hasil = cek_klc('https://klc.kemenkeu.go.id/pknstan-akuntansi-persediaan-penilaian-persediaan-dengan-nrv-bagian-1-akuntansi-aset-agrikultur/')
    print(hasil)
```
